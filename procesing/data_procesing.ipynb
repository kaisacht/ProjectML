```python
def cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13):
    sum = a0+a1+a2+a3+a4+a5+a6+a7+a8+a9+a10+a11+a12+a13
    return sum
    
```

```python
def create_text(method, request, body):
    if method == 'GET':
        return str(method) + ' ' + str(request)
    return str(method) + ' ' + str(request) + ' ' + str(body)
            
data['text'] = data.apply(lambda x: create_text(x['request_http_method'], x['request_http_request'], x['request_body']), axis = 1)
data
```

```python

data = data[data['request_http_request'] != '/']
```

```python
def count_duplicate(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13):
    if a0 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '000 - Normal'
    elif a1 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '272 - Protocol Manipulation'
    elif a2 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '242 - Code Injection'
    elif a3 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '88 - OS Command Injection'
    elif a4 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '126 - Path Traversal'
    elif a5 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '66 - SQL Injection'
    elif a6 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '16 - Dictionary-based Password Attack'
    elif a7 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '310 - Scanning for Vulnerable Software'
    elif a8 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '153 - Input Data Manipulation'
    elif a9 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '248 - Command Injection'
    elif a10 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '274 - HTTP Verb Tampering'
    elif a11 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '194 - Fake the Source of Data'
    elif a12 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '34 - HTTP Response Splitting'
    elif a13 == 1 and cout_attack(a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13)==1:
        return '33 - HTTP Request Smuggling'
    return 'Error'

        
data['label'] = data.apply(lambda x: count_duplicate(x['000 - Normal'],x['272 - Protocol Manipulation'], x['242 - Code Injection'],x['88 - OS Command Injection'],x['126 - Path Traversal'],x['66 - SQL Injection'],x['16 - Dictionary-based Password Attack'],x['310 - Scanning for Vulnerable Software'],x['153 - Input Data Manipulation'],x['248 - Command Injection'],x['274 - HTTP Verb Tampering'],x['194 - Fake the Source of Data'],x['34 - HTTP Response Splitting'],x['33 - HTTP Request Smuggling']), axis = 1)
```
